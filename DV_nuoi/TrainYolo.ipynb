```python

import os
import torch
from ultralytics import YOLO

# Kiểm tra GPU
print("CUDA available:", torch.cuda.is_available())
print("CUDA device count:", torch.cuda.device_count())

# Đường dẫn đến thư mục chứa file data.yaml (thay ROOT_DIR bằng đường dẫn thực tế)
ROOT_DIR = "/content/gdrive/MyDrive/AI/dat"  # Thay đổi cho phù hợp

# Tải mô hình YOLOv8 (bản nhỏ, có thể đổi sang "yolov8m.pt" hoặc "yolov8l.pt" nếu cần hiệu suất tốt hơn)
model = YOLO("yolov8n.pt")  # Load pre-trained model

# Huấn luyện mô hình
results = model.train(
    data=os.path.join(ROOT_DIR, "data.yaml"),
    epochs=200,
    imgsz=640,
    device=0 if torch.cuda.is_available() else "cpu"  # Sử dụng GPU nếu có, nếu không sẽ dùng CPU
)


```
